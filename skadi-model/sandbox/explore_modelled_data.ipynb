```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read the data
file_path = "/Users/adelezaini/Desktop/GEO9300-geophysical_data_science/GEO9300-Gap-Busters/data/modelled-Tuddal-dataset/Tuddal_model_output_fluxes.txt"  # Replace with the path to your text file

# Read the file, assuming columns are space-separated and without headers
df = pd.read_csv(file_path, sep=r'\s+', header=None)

# Assign sequential column names
columns = ['year', 'month', 'day', 'hour'] + [f'col{i}' for i in range(1, len(df.columns) - 4 + 1)]
df.columns = columns

# Step 2: Combine year, month, day, and hour into a datetime index
df['datetime'] = pd.to_datetime(df[['year', 'month', 'day', 'hour']])
df.set_index('datetime', inplace=True)

df = df[df.index.year == 2019]

# Step 3: Drop the original year, month, day, and hour columns
df = df.drop(columns=['year', 'month', 'day', 'hour'])

# Step 4: Plot the data
for column in df.columns:
    plt.figure(figsize=(10, 6))
    plt.plot(df.index, df[column], label=column)
    plt.title(f"Plot of {column}")
    plt.xlabel("Time")
    plt.ylabel("Value")
    plt.grid()
    plt.legend()
    plt.show()
```
