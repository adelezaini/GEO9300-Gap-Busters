```python
path='../data/synthetic_dataset/*.csv'
path_folder='../data/synthetic_dataset/'
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import glob
import matplotlib.cm as cm
import numpy as np

# Define file paths
file_paths = glob.glob(path)

# Initialize a dictionary to store datasets
datasets = {}

# Load each file into a pandas DataFrame
for file_path in file_paths:
    file_name = file_path.split('/')[-1].replace('.csv', '')
    datasets[file_name] = pd.read_csv(file_path)#, parse_dates=['time'])

# Sort datasets alphabetically by key (file name)
datasets = dict(sorted(datasets.items()))
sorted_datasets=datasets
```

```python
# Generate unique colors for each dataset
colors = cm.viridis(np.linspace(0, 1, len(datasets)))

# Plotting the datasets
plt.figure(figsize=(15, 30))

for i, (name, df) in enumerate(datasets.items(), start=1):
    plt.subplot(len(datasets), 1, i)
    df.plot(ax=plt.gca(), color=colors[i - 1], legend=False)  # Assign a unique color
    plt.title(name, fontsize=12)
    plt.xlabel('Index', fontsize=10)
    plt.ylabel(name, fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Determine number of rows and columns for subplots
num_datasets = len(sorted_datasets)
num_cols = 3
num_rows = (num_datasets + 1) // num_cols  # Ensure enough rows to fit all datasets

# Plotting the datasets
fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 2 * num_rows))
axes = axes.flatten()  # Flatten the 2D array of axes for easy iteration

for i, (name, df) in enumerate(sorted_datasets.items()):
    ax = axes[i]
    df.plot(ax=ax, color=colors[i], legend=False)  # Assign a unique color
    ax.set_title(name, fontsize=12)
    #ax.set_xlabel('Index', fontsize=10)
    ax.set_ylabel(name, fontsize=10)

# Hide any unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the required datasets
le_df = pd.read_csv(path_folder+'LE.csv', parse_dates=['time'])  # Replace with the correct path for LE.csv
tair_df = pd.read_csv(path_folder+'Tair.csv', parse_dates=['time'])  # Replace with the correct path for Tair.csv
ts_df = pd.read_csv(path_folder+'Ts.csv', parse_dates=['time'])  # Update this path to your 'Ts.csv' file
```

```python

# Plot 'LE.csv'
plt.figure(figsize=(10, 5))
plt.plot(le_df['time'], le_df['LE'], color='maroon')
plt.title('LE', fontsize=14)
#plt.xlabel('Time', fontsize=12)
plt.ylabel('LE (W/m2)', fontsize=12)
plt.grid(True)
plt.gcf().autofmt_xdate()  # Automatically format datetime x-axis
plt.show()

# Plot 'Ts.csv' vs 'Tair.csv'
plt.figure(figsize=(10, 5))
plt.plot(ts_df['time'], ts_df['Ts'], label='Ts', color='goldenrod')
plt.plot(tair_df['time'], tair_df['Tair'], label='Tair', color='royalblue')
plt.title('Ts vs Tair', fontsize=14)
plt.ylabel('Temperature (K)', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True)
plt.gcf().autofmt_xdate()  # Automatically format datetime x-axis
plt.show()
```
