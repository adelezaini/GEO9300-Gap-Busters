```python
import xarray as xr
import matplotlib.pyplot as plt

# Open the NetCDF file
file_path = "../data/met_analysis_1_0km_nordic_v3_yr_20161108_20240528_formatSURFEXnewTuddal.nc"
data = xr.open_dataset(file_path)

# Display the dataset’s information
data
```

```python
# Plot each variable that has a time dimension
variables = [var for var in data.variables if 'time' in data[var].dims]

# Plot each variable in a separate figure
for var in variables:
    plt.figure(figsize=(10, 4))
    data[var].plot()
    plt.title(f"Time Series of {var}")
    plt.xlabel("Time")
    plt.ylabel(var)
    plt.grid()
    plt.show()
```

```python
subset = data.sel(time=slice("2019-01-01", "2020-12-31"))

# List of variables with a time dimension
variables = [var for var in subset.variables if 'time' in subset[var].dims]

# Plot each variable in a separate figure
for var in variables:
    plt.figure(figsize=(10, 4))
    subset[var].plot()
    plt.title(f"Time Series of {var}")
    plt.xlabel("Time")
    plt.ylabel(var)
    plt.grid()
    plt.show()
```
